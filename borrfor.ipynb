## 1. Contexto de la Actividad
El supermercado de barrio "Buen Grano" vende paquetes de arroz de 1 kg. El administrador se enfrenta a dos costos que se comportan de forma distinta según la cantidad de paquetes $Q$ que decida pedir al proveedor en cada viaje:

1. **Costo de Almacenamiento (Función Lineal):** Entre más arroz pide de una sola vez, más espacio ocupa en la bodega y más cuesta mantenerlo guardado. Este costo sigue la función: 
   $$C_A(Q) = 500Q$$

2. **Costo de Pedido (Función Racional / Proporcionalidad Inversa):** Hacer un pedido implica pagar transporte y logística. Si pide mucho arroz en un solo viaje, hace pocos viajes al año (ahorrando dinero). Si pide poquito, gasta mucho haciendo muchos viajes. El administrador estima que este costo es: 
   $$C_P(Q) = \frac{800,000}{Q}$$

**El Costo Total ($CT$)** que paga el supermercado es la suma de ambos: 
$$CT(Q) = C_A(Q) + C_P(Q)$$

El objetivo logístico es encontrar el tamaño del pedido ($Q$) que minimice el Costo Total.

## 3. Simulación Interactiva de Costos
A continuación, utilizaremos Python para simular cómo se comportan las funciones de costos. 

**Instrucciones:** Ejecuta la siguiente celda de código (puedes presionar `Shift + Enter`). Aparecerá un gráfico con un **deslizador (slider)**. Úsalo para cambiar el tamaño del pedido ($Q$) y observa cómo cambian los costos en tiempo real.

```python
# Importamos las librerías necesarias para matemáticas y gráficos interactivos
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
import warnings
warnings.filterwarnings('ignore')

def simulacion_inventario(Q_seleccionado=20):
    # Definimos el rango de Q (de 10 a 100 unidades)
    Q = np.linspace(10, 100, 200)
    
    # Funciones matemáticas
    C_A = 500 * Q
    C_P = 800000 / Q
    CT = C_A + C_P
    
    # Configuración del gráfico
    plt.figure(figsize=(10, 6))
    plt.plot(Q, C_A, label='Costo de Almacén ($500Q$)', color='#3498db', linewidth=2)
    plt.plot(Q, C_P, label='Costo de Pedido ($800,000/Q$)', color='#e67e22', linewidth=2)
    plt.plot(Q, CT, label='Costo Total ($C_A + C_P$)', color='#2ecc71', linewidth=3)
    
    # Cálculos para el punto seleccionado en el slider
    CA_val = 500 * Q_seleccionado
    CP_val = 800000 / Q_seleccionado
    CT_val = CA_val + CP_val
    
    # Dibujar el punto y la línea vertical del Q seleccionado
    plt.axvline(x=Q_seleccionado, color='red', linestyle='--', alpha=0.6)
    plt.scatter([Q_seleccionado]*3, [CA_val, CP_val, CT_val], color='red', s=80, zorder=5)
    
    # Cuadro de texto con los resultados en tiempo real
    info_text = (
        f"DECISIÓN ACTUAL:\n"
        f"Cantidad (Q) = {Q_seleccionado} paquetes\n"
        f"-------------------------\n"
        f"Costo Almacén = ${CA_val:,.0f}\n"
        f"Costo Pedido  = ${CP_val:,.0f}\n"
        f"COSTO TOTAL   = ${CT_val:,.0f}"
    )
    plt.text(65, 55000, info_text, bbox=dict(facecolor='white', edgecolor='gray', alpha=0.9), fontsize=11)
    
    # Detalles estéticos del gráfico
    plt.title('Simulación: Comportamiento de los Costos de Inventario', fontsize=14, fontweight='bold')
    plt.xlabel('Tamaño del Pedido (Q)', fontsize=12)
    plt.ylabel('Costo en Pesos ($)', fontsize=12)
    plt.ylim(0, 100000)
    plt.xlim(10, 100)
    plt.legend(loc='upper center')
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.show()

# Creamos el deslizador (slider) interactivo
interact(simulacion_inventario, 
         Q_seleccionado=IntSlider(min=10, max=100, step=5, value=20, description='Mover Q:'));
```
