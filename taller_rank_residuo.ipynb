### Recordatorio: Teorema de Rouché–Frobenius

Dado el sistema $A\mathbf{x} = \mathbf{b}$ con $A$ de tamaño $m \times n$:

- Si $\text{rg}(A) \neq \text{rg}([A|\mathbf{b}])$ → **No tiene solución**
- Si $\text{rg}(A) = \text{rg}([A|\mathbf{b}]) = n$ → **Solución única**
- Si $\text{rg}(A) = \text{rg}([A|\mathbf{b}]) < n$ → **Infinitas soluciones**

### Residuo

Si el sistema tiene solución única $\mathbf{x}$, el **residuo** se define como:

$$\mathbf{r} = \mathbf{b} - A\mathbf{x}$$

En aritmética exacta $\mathbf{r} = \mathbf{0}$, pero numéricamente puede ser muy pequeño (del orden de $10^{-15}$). Su norma $\|\mathbf{r}\|$ indica qué tan bien se satisface el sistema.
