## 1. Contexto de la Actividad
Una microempresa llamada **EcoCafe** quiere proyectar su desempeno mensual para decidir si puede expandirse el proximo ano.

El equipo financiero propone dos modelos sencillos:

1. **Ingresos mensuales (sucesion geometrica):** si las ventas crecen a una tasa fija mensual $r$, entonces
   $$I_n = I_0(1+r)^n$$
   donde $I_0 = 2{,}000{,}000$ pesos y $n$ es el numero de meses desde hoy.

2. **Costos mensuales (sucesion aritmetica):** los costos suben una cantidad fija cada mes por inflacion y logistica:
   $$C_n = C_0 + dn$$
   donde $C_0 = 900{,}000$ pesos y $d = 120{,}000$ pesos/mes.

La **utilidad mensual** sera:
$$U_n = I_n - C_n$$

El objetivo es analizar el comportamiento de estas sucesiones y tomar decisiones economicas con base en ellas.

## 3. Simulacion Interactiva de Sucesiones
Ejecuta la siguiente celda y usa el deslizador para explorar como cambian ingresos, costos y utilidad segun el mes y la tasa de crecimiento.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, FloatSlider
import warnings
warnings.filterwarnings('ignore')

def simulacion_sucesiones(mes=12, tasa=4.0, incremento_costos=120000):
    # Parametros base del problema
    I0 = 2000000
    C0 = 900000
    r = tasa / 100

    # Horizonte de analisis
    n = np.arange(0, 37)

    # Sucesiones
    ingresos = I0 * (1 + r) ** n
    costos = C0 + incremento_costos * n
    utilidad = ingresos - costos

    # Valores en el mes seleccionado
    i_mes = I0 * (1 + r) ** mes
    c_mes = C0 + incremento_costos * mes
    u_mes = i_mes - c_mes

    # Buscar primer mes con utilidad objetivo
    objetivo = 2500000
    indices_objetivo = np.where(utilidad >= objetivo)[0]
    primer_mes_obj = int(indices_objetivo[0]) if len(indices_objetivo) > 0 else None

    plt.figure(figsize=(11, 6))
    plt.plot(n, ingresos, label='Ingresos I_n (geometrica)', color='#1f77b4', linewidth=2.5)
    plt.plot(n, costos, label='Costos C_n (aritmetica)', color='#d62728', linewidth=2.5)
    plt.plot(n, utilidad, label='Utilidad U_n = I_n - C_n', color='#2ca02c', linewidth=3)

    # Referencias visuales
    plt.axvline(x=mes, color='gray', linestyle='--', alpha=0.7)
    plt.axhline(y=objetivo, color='purple', linestyle=':', alpha=0.7, label='Meta utilidad: $2,500,000')
    plt.scatter([mes, mes, mes], [i_mes, c_mes, u_mes], color='black', s=55, zorder=6)

    # Cuadro informativo
    objetivo_txt = f'Mes meta utilidad: {primer_mes_obj}' if primer_mes_obj is not None else 'Meta no alcanzada en 36 meses'
    info = (
        f'MES ANALIZADO: {mes}\n'
        f'Tasa crecimiento: {tasa:.1f}% mensual\n'
        f'Ingresos: ${i_mes:,.0f}\n'
        f'Costos:   ${c_mes:,.0f}\n'
        f'Utilidad: ${u_mes:,.0f}\n'
        f'{objetivo_txt}'
    )
    plt.text(20, max(ingresos) * 0.55, info, fontsize=10.5,
             bbox=dict(facecolor='white', edgecolor='gray', alpha=0.92))

    plt.title('Simulacion de Sucesiones en un Contexto Economico', fontsize=14, fontweight='bold')
    plt.xlabel('Mes n', fontsize=12)
    plt.ylabel('Pesos colombianos', fontsize=12)
    plt.xlim(0, 36)
    plt.ylim(min(costos.min(), utilidad.min()) * 1.1, max(ingresos.max(), utilidad.max()) * 1.08)
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend(loc='upper left')
    plt.show()

interact(
    simulacion_sucesiones,
    mes=IntSlider(min=0, max=36, step=1, value=12, description='Mes:'),
    tasa=FloatSlider(min=0.0, max=10.0, step=0.5, value=4.0, description='Tasa %:'),
    incremento_costos=IntSlider(min=50000, max=250000, step=10000, value=120000, description='Delta C:')
);
```
